```python
import pandas as pd
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
data = pd.read_csv('Destination Recommendations - UBS_3_transposed.csv')
```

```python
data.head()
```

```python
data.dtypes
```

```python
data = data.drop(labels='Destinations',axis=1)
```

```python
data.head()
```

# Run the apriori algorithm

```python
frequent_itemsets = apriori(data,0.2,True)
```

```python
frequent_itemsets
```

```python
type(frequent_itemsets)
```

```python
frequent_itemsets.sort_values('support',ascending=0)
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.4)
```
